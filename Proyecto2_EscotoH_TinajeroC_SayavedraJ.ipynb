## Empresa AMD: ##

```python
import numpy as np
import pandas as pd
import pandas_datareader as data
import matplotlib.pyplot as plt
%matplotlib inline
def load_adj_close_AMD(ticker, data_source, start_date, end_date):
    panel_data = data.DataReader(ticker, data_source, start_date, end_date)
    closes_AMD = panel_data.loc['Adj Close']
    all_weekdays = pd.date_range(start=start_date, end=end_date)
    closes_AMD = closes_AMD.reindex(all_weekdays)
    closes_AMD = closes_AMD.fillna(method='ffill')
    return closes_AMD
ticker = ['AMD']
data_source = 'yahoo'
start_date = '2015-01-04'
end_date = '2016-12-31'
closes_AMD = load_adj_close_AMD(ticker, data_source, start_date, end_date)
closes_AMD.plot(figsize=(8,6));
```

**Rendimientos diarios AMD:**

```python
def mu_std_daily_ret_AMD(closes_AMD):
    daily_returns_AMD = (np.log(closes_AMD/closes_AMD.shift(1)))[1:]
    mu_AMD = daily_returns_AMD.mean().values[0]
    sigma_AMD = daily_returns_AMD.std().values[0]
    return daily_returns_AMD, mu_AMD, sigma_AMD
daily_returns_AMD, mu_AMD, sigma_AMD = mu_std_daily_ret_AMD(closes_AMD)
mu_AMD, sigma_AMD
```

```python
daily_returns_AMD.plot(figsize=(8,6));
```
