```python
import random

def todojuego():
    def game1():
        m=['-']*10
        abc=[chr(i) for i in range(64,91)]
        coord={'A':1,'B':2,'C':3,'D':4, 'E':5, 'F':6, 'G':7, 'H':8, 'I':9}
        num=0
        for fil in range(10):
            m[fil]=['-']*10
            for col in range(10):
                m[0][col]=abc[col]
        for i in range(10):
            m[i][0],m[0][0]=str(num),' '
            num+=1
        return m,coord
    mapa,letra=game1()
    print('=====================','\n','MAPA CON COORDENADAS'+'\n'+'=====================')
    for i in mapa:
        print(' ','  '.join(i))
 
    def fichas():
        con=0
        while con<40:
            mapa[random.randint(1,9)][random.randint(1,9)]='X'
            con+=1
        return mapa
    fichas()
    def jugadas():
        print('\n'+'TIENES 15 INTENTOS, PARA HUNDIR 5 BARCOS')
        intento=0
        elim=0
        while intento<15:
            intento+=1
            #cambia los valores de 'fila' y 'mun' para ingresarlos manualmente
            fila=random.randint(1,9)
            num=(random.randint(65,73))
            #fila=input('INGRESE FILA: ')
            #num=int(input('INGRESE LETRA (65-73): '))
            colu=letra[chr(num)]
            print('intento',intento,'Posición('+str(fila)+','+chr(num)+')',end=' ___ ')
            if mapa[fila][colu]=='X':
                mapa[fila][colu]='O'
                print('Barco hundido')
                elim+=1
            else:
                print('Agua')
        return mapa,elim
 
    final,remo=jugadas()
    if remo>=5:
        print('\n','FELICIDADES, ACABAS DE HUNDIR:',remo,'BARCOS')
    else:
        print('\n','PERDISTE, SOLO ELIMINASTE:',remo,'BARCOS')
    for i in final:
        print(' ','  '.join(i))
letra=[chr(i) for i in range(65,91)]
n='A'
while n in letra:
    print()
    n=input('PRESIONA UNA LETRA INICIAR EL JUEGO: ').upper()
    if n in letra:
        todojuego()
```
